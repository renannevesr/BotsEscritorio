```python
import pandas as pd
df_remuneracao = pd.read_csv('202401_Remuneracao.csv', encoding="ISO-8859-1",sep=";")
df_pensionista = pd.read_csv('PENSIONISTAS_012024.csv', encoding="ISO-8859-1",sep=";")
```

```python
df_filtrado_pensionista = df_pensionista.loc[:, ['NOME DO SERVIDOR INSTITUIDOR', 'CPF SERVIDOR INSTITUIDOR', 'NOME DO BENEFICIARIO', 'CPF DO BENEFICIARIO','CARGO SERVIDOR INSTITUIDOR','TIPO DE BENEFICIARIO','RENDIMENTO BRUTO','RENDIMENTO LIQUIDO']]
df_filtrado_pensionista = df_filtrado_pensionista.dropna(subset=['TIPO DE BENEFICIARIO'])
```

```python
df_remuneracao_filtrado= df_remuneracao.loc[:, [ 'CPF', 'NOME','IRRF (R$)', 'REMUNERAÇÃO BÁSICA BRUTA (R$)']]
```

```python
def formatar_numero(numero):
    numero = ''.join(filter(str.isdigit, numero))  
    numero_formatado = "***.{:03d}.{}-**".format(int(numero[:-3]), int(numero[-3:]))
    return numero_formatado

# Aplicando a função à coluna 'numero'
df_filtrado_pensionista['CPF DO BENEFICIARIO'] = df_filtrado_pensionista['CPF DO BENEFICIARIO'].apply(formatar_numero)
```

```python
df_filtrado = df_filtrado_pensionista[df_filtrado_pensionista['TIPO DE BENEFICIARIO'].str.contains("INV")]
```

```python
result = pd.merge(df_remuneracao_filtrado, df_filtrado, on=["CPF DO BENEFICIARIO", "NOME DO BENEFICIARIO"])
```
